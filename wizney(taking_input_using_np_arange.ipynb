```python
import numpy as np
import matplotlib.pyplot as plt

# input signal
signal = np.arange(5)  # A simple 1D signal from 0 to 4

# filter
fltr = np.arange(1, 4)

# signal = [1,2,3,1]
# fltr = [1,2,1,-1]

#calculating length
signal_length = len(signal)
fltr_length = len(fltr)
output_length = signal_length + fltr_length - 1
output = np.zeros(output_length)

# convolution
for i in range(output_length):
    for j in range(max(0, i - fltr_length + 1), min(i + 1, signal_length)):
        output[i] += signal[j] * fltr[i - j]

# Display the results
print("Signal:")
print(f"Signal: {signal}")

print("\nFilter:")
print(f"Filter: {fltr}")

print("\nConvolution Output:")
print(f"Output: {output}")

#plot input signal, filter & convolution output
plt.figure(figsize=(10, 6))

#plot input signal
plt.subplot(3, 1, 1)
plt.stem(signal, basefmt=" ")
plt.title('Input Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axhline(0, color='k', linestyle='--')
plt.axvline(0, color='k', linestyle='--')
plt.grid(True)

#plot filter
plt.subplot(3, 1, 2)
plt.stem(fltr, basefmt=" ")
plt.title('Filter')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axhline(0, color='k', linestyle='--')
plt.axvline(0, color='k', linestyle='--')
plt.grid(True)

#plot convolution output
plt.subplot(3, 1, 3)
plt.stem(output, basefmt=" ")
plt.title('Convolution Output')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axhline(0, color='k', linestyle='--')
plt.axvline(0, color='k', linestyle='--')
plt.grid(True)

plt.tight_layout()
plt.show()

```
